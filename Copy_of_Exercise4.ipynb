# Exercise 4

## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
```

## Load the data

```python
X = pd.read_csv('creditcard.csv')

# Dropping the CUST_ID column from the data
X = X.drop('CUST_ID', axis = 1)

# Handling the missing values
X.fillna(method ='ffill', inplace = True)

print(X.head())
```

## Pre-process the data

```python
# Scaling the data to bring all the attributes to a comparable level
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Normalizing the data so that
# the data approximately follows a Gaussian distribution
X_normalized = normalize(X_scaled)

# Converting the numpy array into a pandas DataFrame
X_normalized = pd.DataFrame(X_normalized)
```

## Reduce the dimensionality of the data using PCA

```python
pca = PCA(n_components = 2)
X_principal = pca.fit_transform(X_normalized)
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2']
print(X_principal.head())
```

## Train a DBSCAN model using default values

```python
# Numpy array of all the cluster labels assigned to each data point
db_default = DBSCAN().fit(X_principal)
labels = db_default.labels_
```

## Plot the clustered data

```python
# Building the label to colour mapping
colours = {}
colours[0] = 'r'
colours[1] = 'g'
colours[2] = 'b'
colours[-1] = 'k'

# Building the colour vector for each data point
cvec = [colours[label] for label in labels]

# For the construction of the legend of the plot
r = plt.scatter(X_principal['P1'], X_principal['P2'], color ='r');
g = plt.scatter(X_principal['P1'], X_principal['P2'], color ='g');
b = plt.scatter(X_principal['P1'], X_principal['P2'], color ='b');
k = plt.scatter(X_principal['P1'], X_principal['P2'], color ='k');

# Plotting P1 on the X-Axis and P2 on the Y-Axis
# according to the colour vector defined
plt.figure(figsize =(9, 9))
plt.scatter(X_principal['P1'], X_principal['P2'], c = cvec)

# Building the legend
plt.legend((r, g, b, k), ('Label 0', 'Label 1', 'Label 2', 'Label -1'))

plt.show()
```

1. Train a DBSCAN model

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN, KMeans
from sklearn.metrics import silhouette_score

# Step 1: Preprocess the data
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_principal)

# Reduce dimensions using PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

# Step 2: Optimize DBSCAN (fine-tuning `eps` and `min_samples`)
best_dbscan_score = -1
best_dbscan_labels = None
best_eps = None
best_min_samples = None

for eps in np.linspace(0.05, 0.4, 15):  # Finer range for `eps`
    for min_samples in range(3, 25, 2):  # Finer range for `min_samples`
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(X_reduced)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters > 1:
            silhouette_avg = silhouette_score(X_reduced, labels)
            if silhouette_avg > best_dbscan_score:
                best_dbscan_score = silhouette_avg
                best_dbscan_labels = labels
                best_eps = eps
                best_min_samples = min_samples

# Step 3: Optimize KMeans (for fallback)
best_kmeans_score = -1
best_kmeans_labels = None
best_k = None

for k in range(2, 12):  # Expand k range for more cluster testing
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_reduced)
    silhouette_avg = silhouette_score(X_reduced, labels)
    if silhouette_avg > best_kmeans_score:
        best_kmeans_score = silhouette_avg
        best_kmeans_labels = labels
        best_k = k

# Step 4: Evaluate and select the best method
if best_dbscan_score > best_kmeans_score:
    print(f"Using DBSCAN: eps={best_eps}, min_samples={best_min_samples}")
    print(f"Silhouette Score: {best_dbscan_score:.2f}")
    final_labels = best_dbscan_labels
    final_score = best_dbscan_score
else:
    print(f"Using KMeans: k={best_k}")
    print(f"Silhouette Score: {best_kmeans_score:.2f}")
    final_labels = best_kmeans_labels
    final_score = best_kmeans_score

# Step 5: Visualize the best clustering
plt.figure(figsize=(8, 8))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=final_labels, cmap='viridis', s=30)
plt.title(f'Best Clustering (Silhouette Score: {final_score:.2f})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar()
plt.show()
```
